### Ollama Embeddings

Ollama is open source.
Run Llama 3, Phi 3, Mistral, Gemma and other models.

Install Ollama in the machine

1. Open cmd
2. ollama run gemma:2b     <!-- If this model is not downloaded, it will download and then use the model. It is around 1.4GB -->
3. Perform query on the model and will recieve the response.

Similarly, can use any model of interest.

Ollama supports embedding model, making it possible to build RAG app that combine text prompts with existing documents or other data

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma:2b")      # By default it uses Llama2
)

embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first character of the Greek alphabet",
        "Beta is the second character of the Greek alphabet",
    ]
)

r1
```

```python
r1[0]
```

```python
len(r1[0])
```

```python
len(r1)
```

```python
embeddings.embed_query("What is the second letter of Greek alphabet")
```

```python
# Other embedding models
# https://ollama.com/blog/embedding-models

embeddings = OllamaEmbeddings(model="mxbai-embed-large")

text = "This is a test document"
query_result = embeddings.embed_query(text=text)
query_result
```
